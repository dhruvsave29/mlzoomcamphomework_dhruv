```python
u = np.array ([1, 2, 3])
v = np.array ([4, 5, 6])
u + v
```

```python
def vector_vector_multiply(u, v):
    assert u.shape[0] == v.shape[0]

    n =u.shape[0]
    result = 0
    for i in range(n):
        result = result +u[i] * v[i]
    return result
```

```python
vector_vector_multiply(u, v)
```
